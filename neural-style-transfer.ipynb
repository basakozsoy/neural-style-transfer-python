## Simple Neural Style Transfer using TensorFlow Hub

```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
#load a neural style transfer model from tensorflow hub
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
#load image
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3) #make sure the image has 3 channels
    img = tf.image.convert_image_dtype(img, tf.float32) #convert the values to float32
    img = img[tf.newaxis, :] #put inside a new array
    return img
```

```python
#input
base_image = load_image('country.jpg')
style_image = load_image('vangogh.jfif')
```

```python
#base image
plt.imshow(np.squeeze(base_image))
plt.show()
```

```python
#style image
plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
#convert the inputs to tf.constant
stylized_image = model(tf.constant(base_image), tf.constant(style_image))[0] 
plt.imshow(np.squeeze(stylized_image))
plt.show()
```

```python
#saving the image with BGR to RGB conversion, multiply by 255 to reverse the normalization
cv2.imwrite('stylized_image.jpg', cv2.cvtColor(np.squeeze(stylized_image)*255, cv2.COLOR_BGR2RGB))
```
